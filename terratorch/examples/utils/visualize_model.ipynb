### Visualize a TerraTorch Model

This notebooks demonstrates how you can visualize the architecture of any TerraTorch backbone.

Note: The visualization requires Graphviz to be installed on your system; e.g. on macOS/Linux: `brew install graphviz`

```python
!pip install torchview
!pip install graphviz
```

```python
import torch
from terratorch.registry import BACKBONE_REGISTRY
from terratorch.utils import view_api
```

```python
# Get model and sample input
model = BACKBONE_REGISTRY.build("prithvi_eo_v2_300")
trial_data = torch.zeros(1, 6, 224, 224)
```

```python
view_api(model, input_data=trial_data, config={"expand_nested":True, "depth":2, "graph_dir": "LR"})
```
